```python
import pandas as pd

def get_matrix(path):
    df = pd.read_csv(path, header=None)

    x, y, n = [], [], []
    for _, row in df.iterrows():
        n.append(int(row[1]))
        strings = row[0].split()
        x.append(strings[-1][0:-1])
        y.append(strings[-3])

    names = set(x)
    names = sorted(names)
    
    matrix_dict = {name_x: {name_y: 0 for name_y in names} for name_x in names}
    # Populate the nested dictionary with values from x, y, and n
    for i in range(len(x)):
        matrix_dict[x[i]][y[i]] = n[i]
        
    matrix = pd.DataFrame(matrix_dict)
    return matrix

matrix = get_matrix(f"../examples/pairwise/response_davinci-002.csv")
matrix.to_csv(f"../examples/pairwise/response_davinci-002_matrix.csv")
```

```python
for i in range(1, 21):
    matrix = get_matrix(f"../examples/pairwise/exp_20_temp_0.7/response_davinci-003_{i}.csv")
    matrix.to_csv(f"../examples/pairwise/exp_20_temp_0.7/exp_20_temp_0.7_matrix/response_davinci-003_matrix_{i}.csv")
```

```python
dfs = []
for i in range(1, 21):
    df = pd.read_csv(f"../examples/pairwise/exp_20_temp_0.7/exp_20_temp_0.7_matrix/response_davinci-003_matrix_{i}.csv")
    df = df.iloc[:, 1:]
    df = df.apply(pd.to_numeric)
    dfs.append(df)

sum_df = pd.DataFrame(0, index=dfs[0].index, columns=dfs[0].columns)

# Sum the values in all DataFrames
for df in dfs:
    sum_df += df

# Calculate the mean by dividing the sum by the number of DataFrames
mean_df = sum_df / len(dfs)
mean_df.index = mean_df.columns
mean_df = mean_df.sort_index(axis=0).sort_index(axis=1)
mean_df.to_csv(f"../examples/pairwise/response_davinci-003_0.7_mean_matrix.csv")
mean_df
```
